```python
import pandas as pd 
import matplotlib.pyplot as plt 
import random

df_student = pd.read_csv("dataset.csv", delimiter=',', header='infer')

df_student.head()
```

```python
df_student.dtypes
```

```python
counts = df_student.Gender.value_counts()
plt.pie(counts, labels=counts.index, autopct='%1.1f%%') 
plt.legend()

plt.title("Gender Ratio")
plt.show()
```

```python
counts = df_student.Grade.value_counts()
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', explode=[random.uniform(0, 0.1) for _ in range(4)]) 
plt.legend()

plt.title("Grade")
plt.show()

```

```python

plt.figure(figsize=(12,4.8))   
plt.bar(df_student.Name, df_student.Marks, label="marks", zorder=2)

plt.title("Marks of each student")
plt.ylabel("Marks")
plt.xlabel("Name")
plt.ylim(top=101)
plt.xticks(rotation=90)
plt.grid(zorder=0)
plt.legend()

plt.show()
```

```python
import numpy as np

x = np.arange(df_student.Name.count())   # [0,1,2,3]
width = 0.35

plt.bar(x - width/2, df_student.StudyHours, width, label='Study Hours')
plt.bar(x + width/2, df_student.Marks, width, label='Marks')

plt.xticks(x, df_student.Name)
plt.xlabel("Name")
plt.ylabel("Marks")
plt.title("Marks vs StudyHour")
plt.xticks(rotation=90)
plt.legend()

plt.show()
```

```python
marks = df_student.groupby("Gender")["Marks"].mean()
studyhour = df_student.groupby("Gender")["StudyHours"].mean()
hardiksassignment = df_student.groupby("Gender")['Attendance'].mean()

x = np.arange(marks.count())
width = 0.25

plt.bar(x-width, marks.values, width, label="Marks")
plt.bar(x, studyhour.values, width, label="Study Hours")
plt.bar(x + width, hardiksassignment.values, width, label="Attendance")

plt.xticks(x, marks.index)
plt.grid(axis='y', zorder=0)
plt.legend()
plt.title("gender-wise average distribution")

plt.show()
```

```python
studyhour = df_student.groupby("Grade")["StudyHours"].mean()
hardiksassignment = df_student.groupby("Grade")['Attendance'].mean()

x = np.arange(marks.count())
width = 0.25

plt.bar(x - width/2, studyhour.values, width, label="Study Hours")
plt.bar(x + width/2, hardiksassignment.values, width, label="Attendance")

plt.xticks(x, marks.index)
plt.grid(axis='y', zorder=0)
plt.legend()
plt.title("grade-wise average distribution")

plt.show()
```
