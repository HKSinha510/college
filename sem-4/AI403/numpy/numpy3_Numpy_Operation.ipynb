# element-wise operation
the operation is applied to each corresponding element of the array. no loops are required

```python
import numpy as np
a = np.array([2, 3, 4])
b = np.array([5, 6, 7])

print(a+b)


arr2 = np.array([[1, 2, 3], [2, 3, 4]])
arr3 = np.array([[1, 2, 3], [4, 5, 6]])
print(arr2+arr3)
```

# array scalar operation
scalar is applied to each elements.

```python
print(a * 5)
```

# matrix multiplication


```python
mat1 = np.array([[2, 3], [5, 6]])
mat2 = np.array([[4, 5], [1, 2]])

print(mat1 * mat2)
print(mat1 @ mat2) # matrix multiplicaiton
```

# universal function(unfuncs)
it is a vectorised function that perform element wise operation on array.


```python
np.add(a, b)
np.subtract(a, b)
```

- 2d are compatible if they are equal or one of them is 1d.
- expand dimension if one array has fewer dimension than the other, prepend one to it's shape, until the no. of dimension matches

# Aggrigation funciton
It process an entire array to return a single summary value.

## Common functions are
1. sum()
2. mean()
3. min()
4. max()
5. std(): standard deviation
6. var(): variance
7. prod(): product

```python
arr1 = np.array([2, 4, 7])

#sum
print(np.sum(arr1))

#mean
print(np.mean(arr1))

#min
print(np.min(arr1))

#max
print(np.max(arr1))

#std
print(np.std(arr1))

#var
print(np.var(arr1))

#prod
print(np.prod(arr1))
```
