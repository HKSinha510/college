# 1. Indexing
## 1D idexing



```python
arr1d[1]
arr1d[-2]

arr1d[1:]
arr1d[1:3:1]
arr1d[::2]
```

## 2D indexing

```python
#array[row_index, col_index]

# print(arr2d[0][1])
# print(arr2d[(0, 1)])
arr2d[:,1]

#slicing
#array[row:row, col:col]
arr2d[:2,1:]
arr2d[:,1:2]
arr2d[1:, :2]
```

## 3d indexing

```python
#array[block, row, column]
arr3d[0,0,1]
arr3d[1,0,1]
arr3d[1][0][1]


#slicing
arr3d[:, :, 1]
```
