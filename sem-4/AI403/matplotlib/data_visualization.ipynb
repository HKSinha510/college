#### Line Plot/Line Graph
it shows trends over a continuous variable

##### It is of two types *single line* and *multi line*
1. single line plot one series of data points, connected my lines
2. multi lines compare two of more dataset on the same axis

```python
x = [i for i in range(1, 6)]
y = [i for i in range(10, 51, 10)]

plt.plot(x, y)

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line plot")

plt.show()
```


##### customization
- colors
- linestyle
- markers

```python
plt.plot(x, y, color="red", marker="o", linestyle="--")
plt.show()
```

```python
#multi-line graph

x = [i for i in range(1, 6)]
y1 = [i for i in range(10, 51, 10)]
y2 = [9, 25, 36, 90, 99]

plt.plot(x, y1, color="blue", marker="o", linestyle="--")
plt.plot(x, y2, color="green", marker="x")

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("A graph")

plt.show()
```

##### Legends and grids

- Legends

label agrument in plt.plot()
`plt.plot(x, y, label="<label>")`
`plt.legend()`

- Grid
`plt.grid(True)`

```python
#multi-line graph

x = [i for i in range(1, 6)]
y1 = [i for i in range(10, 51, 10)]
y2 = [9, 25, 36, 90, 99]

plt.plot(x, y1, color="blue", marker="x", linestyle=":", label="IOT")
plt.plot(x, y2, color="green", marker="o", label="AIML")

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("A graph")
plt.legend()
plt.grid(True)

plt.show()
```

#### Bar Plot
It is a graphical representation of data using rectangular bars, where the length and height of each bar is proposnal to the value it represent. <br>
It is mainly used for comparing quantity across different categories.

##### Types of Bar Plot
1. Vertical Bar Plot
2. Horizontal Bar Plot
3. Multiple Bar Chart

##### syntax
`plt.bar(x, height)`

###### 1. Vertical Bar Plot

```python
#Q. prepare a bar chart to show marks scored by a student in 5 subjects

x = ["English", "Maths", "Science", "CS"]
h = [80, 20, 40, 90]

plt.bar(x, h, width=0.1, label="Student A", color=['b', 'c', 'm', 'y'])

plt.xlabel("subjects")
plt.ylabel("marks")
plt.legend()


plt.show()
```

###### 2. Horizontal Bar

```python
# horizontal Bar

x = ["English", "Maths", "Science", "CS"]
h = [80, 20, 40, 90]

plt.barh(x, h, height=0.2, label="Student A", color=['b', 'c', 'm', 'y'])

plt.ylabel("subjects")
plt.xlabel("marks")
plt.legend()


plt.show()
```

##### Labelling each bar

```python

x = ["English", "Maths", "Science", "CS"]
h = [80, 20, 40, 90]

bar_width = 0.9

plt.bar(x, h, width=bar_width, label="Student A", color=['b', 'c', 'm', 'y'])

for i in range(len(x)):
    plt.text(i - (bar_width//2), h[i] + 1, h[i])

plt.xlabel("subjects")
plt.ylabel("marks")
plt.legend()


plt.show()
```
