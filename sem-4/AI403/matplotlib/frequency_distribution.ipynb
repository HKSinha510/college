## Bar Plot

### 1. Single bar plot

```python
import matplotlib.pyplot as plt

x = ['Math', 'Science', 'S. Science', 'Hindi', 'English']
y = [50, 60, 70, 55, 60]

plt.bar(x, y)


plt.show()
```

```python
import numpy as np

subject = ['Math', 'Science', 'S. Science', 'Hindi', 'English']
x = np.arange(len(subject))
y = [50, 60, 70, 55, 60]
y2 = [55, 65, 75, 90, 55]

w = 0.4

plt.bar(x - (w/2), y, width=w, label='term1')
plt.bar(x + (w/2), y2, width=w, label='term2')

plt.xlabel("Subjects")
plt.ylabel("Marks")

plt.legend()


plt.show()
```
