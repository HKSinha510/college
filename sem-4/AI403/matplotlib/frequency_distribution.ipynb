## Bar Plot

### 1. Single bar plot

```python
import matplotlib.pyplot as plt

x = ['Math', 'Science', 'S. Science', 'Hindi', 'English']
y = [50, 60, 70, 55, 60]

plt.bar(x, y)


plt.show()
```

```python
import numpy as np

subject = ['Math', 'Science', 'S. Science', 'Hindi', 'English']
x = np.arange(len(subject))
y = [50, 60, 70, 55, 60]
y2 = [55, 65, 75, 90, 55]

w = 0.4

plt.bar(x - (w/2), y, width=w, label='term1')
plt.bar(x + (w/2), y2, width=w, label='term2')

plt.xlabel("Subjects")
plt.ylabel("Marks")

plt.legend()


plt.show()
```

## Histogram

```python
import math
age=[18, 19, 20, 21, 22, 22, 23, 24, 25, 26, 28, 30, 32, 35, 38, 40, 45, 50, 55, 60, 65, 70]

noOfClass = 1+3.3 * math.log10(len(age))
print(noOfClass)
classWidth = (max(age)-min(age))/noOfClass
print(classWidth)

plt.hist(age, edgecolor='white')

plt.grid()
plt.show()
```

1. What does each bar in the histogram represent?
frequency of each age group

2. Is the distribution symmetric, left-skewed, or right-skewed? Why?
right-skewed

3. Where is the peak (mode) of the distribution?
