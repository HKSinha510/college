# Measure of Dispersion

## 1. range

```python
import matplotlib.pyplot as plt
import numpy as np

data = [12, 7, 25, 3, 18, 10]

min_value = min(data)
max_value = max(data)
range = max_value - min_value

print(np.ptp(data))

plt.plot(data, marker="+")

plt.axhline(max_value, linestyle="--", color="black", label='Minimum Value')
plt.axhline(min_value, linestyle="-.", color="black", label='Maximum Value')
plt.legend()
plt.show()


```

## Quartile Deviation

```python
data = [12, 7, 25, 3, 18, 10, 15, 20]

q1 = np.percentile(data, 25)
q2 = np.percentile(data, 50)
q3 = np.percentile(data, 75)

quartile_deviation = (q3-q1)/2

print(quartile_deviation)


print(np.ptp(data))

plt.boxplot(data)

plt.axhline(min(data), label="minimum value", color="black", linestyle="--")
plt.axhline(max(data), label="maximum value", color="black", linestyle="-.")

plt.axhline(q1, label="lower quartile", linestyle="--")
plt.axhline(q3, label="upper quartile", linestyle="-.")

plt.legend()

plt.show()
```

### left sc

```python
data = [20, 30, 35, 40, 42, 45, 47, 48, 49, 100]

q1 = np.percentile(data, 25)
q2 = np.percentile(data, 50)
q3 = np.percentile(data, 75)

quartile_deviation = (q3-q1)/2

print(quartile_deviation)


print(np.ptp(data))

plt.boxplot(data)

plt.axhline(min(data), label="minimum value", color="black", linestyle=":")
plt.axhline(max(data), label="maximum value", color="black", linestyle="--")

plt.axhline(q1, label="lower quartile", linestyle=":")
plt.axhline(q3, label="upper quartile", linestyle="--")

plt.legend()

plt.show()
```

## 3. Mean Deviation

```python
# from math import *

data = [12, 7, 25, 3, 18, 10, 15, 20]

m = np.mean(data)


mean_deviation = np.mean([abs(x-m) for x in data])
print(mean_deviation)
```

## variance

```python
data = [12, 7, 25, 3, 18, 10, 15, 20]


variance = np.var(data)
print(variance)
```

## standard deviation

```python
data = [12, 7, 25, 3, 18, 10, 15, 20]


standdv = np.std(data)
print(standdv)
```

## coffiicient of variable

```python
cv = standdv/np.mean(data)

print(cv*100)
```

# Measure of shape

```python
from statistics import mode
from scipy.stats import skew
import math

#right skewed
data = [10, 11, 12, 12, 13, 13, 14, 14, 15, 15,16, 16, 17, 18, 19, 22, 24, 26]

m = np.mean(data)
me = np.median(data)
mo = mode(data)
print(skew(data))

plt.figure()
noOfClass = 1+3.3 * math.log10(len(data))
classWidth = (max(data)-min(data))/noOfClass
print(classWidth)
plt.hist(data, bins=round(classWidth))
plt.axvline(m, color="black", linestyle=":", label="Mean")
plt.axvline(me, color="black", linestyle="--", label="Median")
plt.axvline(mo, color="black", label="Mode")
plt.legend()

plt.figure()
plt.boxplot(data)
plt.axhline(m, color="black", linestyle=":", label="Mean")
plt.axhline(me, color="black", linestyle="--", label="Median")
plt.axhline(mo, color="black", label="Mode")
plt.legend()

plt.show()
```
