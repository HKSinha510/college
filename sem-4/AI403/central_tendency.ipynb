# mean

## Mean for uniform data

```python
import matplotlib.pyplot as plt
import numpy as np

data = [65, 70, 72, 68, 75, 73, 71, 69]
mean_value = np.mean(data)
print(mean_value)

plt.hist(data, edgecolor='white', bins=5)
plt.axvline(mean_value, color="black")


plt.show()

```

## Mean for skewed data

```python
data =[20000, 22000, 25000, 23000, 24000, 21000, 500000]
mean_value = np.mean(data)

plt.hist(data, edgecolor="white")
plt.axvline(mean_value, color="black")
```

```python
x = ['Math', 'Science', 'S. Science', 'Hindi', 'English']
y = [50, 60, 70, 55, 60]

mean_value = np.mean(y)
print(mean_value)

plt.bar(x, y)
plt.axhline(mean_value, color="black")

plt.xlabel("subjects")
plt.ylabel("marks")

plt.show()
```
